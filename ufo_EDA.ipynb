```python
import geopy
from geopy.geocoders import Nominatim
import time

def get_coordinates(city, state, cache):
    # Check if the coordinates are already cached
    if (city, state) in cache:
        return cache[(city, state)]
    
    geolocator = Nominatim(user_agent="ufo_sighting_locations")
    location = geolocator.geocode(f"{city}, {state}", timeout=10)
    
    if location:
        # Cache the coordinates to avoid redundant requests
        cache[(city, state)] = (location.latitude, location.longitude)
        return location.latitude, location.longitude
    else:
        return None, None

cache = {}

# Example usage
city = "Lauglintown"
state = "PA"
latitude, longitude = get_coordinates(city, state, cache)
print(f"Latitude: {latitude}, Longitude: {longitude}")
```

#### The Dreaded Duration Column

getting an idea of what the duration texts look like

```python
duration = set()

for duration_len in ufo_data['duration']:
    duration.add(duration_len)

duration
```

Created a couple functions to aid in cleaning/ standardizing the duration texts and converting them in terms of minutes

```python
import pandas as pd
import numpy as np
import re
from quantulum3 import parser

# Sample data: replace with your actual data
duration

# Create a DataFrame (or use your existing DataFrame)
df = pd.DataFrame(list(duration), columns=['duration'])

# Function to convert words to numbers
def word_to_num(word):
    word_map = {
        "one": 1, "two": 2, "three": 3, "four": 4, "five": 5,
        "six": 6, "seven": 7, "eight": 8, "nine": 9, "ten": 10,
        "half": 0.5, "quarter": 0.25, "thirty": 30
    }
    return word_map.get(word.lower(), None)

# Function to clean and standardize the time durations
def standardize_duration(duration):
    if isinstance(duration, str):
        # Lowercase the input to avoid case issues
        duration = duration.lower()
        
        # Remove unwanted characters (e.g., '?', (', ')') keep "-", "/",":", "."
        duration = re.sub(r'[^a-zA-Z0-9\s./\-:]', '', duration)

        duration = re.sub(r'(?<=\d)\.(?=\D)', '', duration)  # Remove dot if it is not between digits
        duration = re.sub(r'(?<=\d)(?=\.\d)', '.', duration)  # Keep the dot if it is between digits
        
        # Convert any word-based numbers into numeric values
        words = duration.split()
        for i, word in enumerate(words):
            num_value = word_to_num(word)
            if num_value is not None:
                words[i] = str(num_value)  # Replace word with its numeric equivalent
        duration = ' '.join(words)
        
        # Handle time ranges like '5-10 minutes'
        if '-' in duration or 'to' in duration or '/' in duration:
            numbers = re.findall(r'\d+\.?\d*', duration)  # Find all numbers (including decimals)
            
            if len(numbers) == 2:  # If there are exactly two numbers (range)
                avg = np.mean([float(numbers[0]), float(numbers[1])])  # Calculate the average
                
                # Convert to minutes if the unit is 'minute' or 'min'
                if 'minute' in duration or 'min' in duration or 'minutes' in duration:
                    return f"{avg:.6f}"  # Return as string with six decimal places
                
                # Convert to minutes if the unit is 'hour' or 'hours'
                if 'hour' in duration or 'hours' in duration or 'hr' in duration or 'hrs' in duration:
                    return f"{avg * 60:.6f}"  # Convert hours to minutes and return as string
                
                # Convert to minutes if the unit is 'second' or 'sec'
                if 'second' in duration or 'sec' in duration or 'secs' in duration:
                    return f"{avg / 60:.6f}"  # Convert seconds to minutes and return as string
                
                return f"{avg:.6f}"  # Return the average value as a string with 6 decimals
        
        # Handle values with time units (minutes or hours)
        if 'minute' in duration or 'min' in duration or 'minutes' in duration:
            minutes = re.findall(r'\d+\.?\d*', duration)
            if minutes:
                return f"{minutes[0]}"
        
        if 'hour' in duration or 'hours' in duration or 'hr' in duration:
            hours = re.findall(r'\d+\.?\d*', duration)
            if hours:
                return f"{float(hours[0]) * 60}"  # Convert to minutes directly
            
        if 'second' in duration or 'sec' in duration or 'secs' in duration:
            second = re.findall(r'\d+\.?\d*', duration)
            if second:
                return f"{float(second[0]) / 60}"  # Convert to minutes
        
        # Handle specific cases (e.g., "one", "half")
        if 'one' in duration:
            return "1 minute"  # One minute
        elif 'half' in duration:
            return "30 minutes"  # Half an hour
        
        # Handle unknown or unparseable cases (return NaN)
        return np.nan

# Function to parse duration string into minutes
def parse_duration_to_minutes(text):
    if isinstance(text, str):
        quantities = parser.parse(text)
        for q in quantities:
            value = q.value
            unit = q.unit.name.lower()
            
            if 'second' in unit:
                return value / 60
            elif 'minute' in unit:
                return value
            elif 'hour' in unit:
                return value * 60
            elif 'day' in unit:
                return value * 1440  # 24 * 60
    return np.nan  # If no recognizable duration found

# Apply the function to clean and standardize durations, then parse to minutes
df['standardized_duration'] = df['duration'].apply(standardize_duration)

# Step 2: Identify rows where 'standardized_duration' is NaN
subset_df = df[df['standardized_duration'].isna()]

# Step 3: Apply the parse_duration_to_minutes function to the subset
subset_df['standardized_duration'] = subset_df['duration'].apply(parse_duration_to_minutes)

# Step 4: Update the original DataFrame with the new values from subset_df
df.update(subset_df)

# Show the result
print(df[['duration', 'standardized_duration']][301:350])
```
